```python
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
df_firms = pd.read_excel("./data/firms.xlsx", sheet_name = "dotcom_firms")
```

```python
## merge with CRSP
df_option_crsp = pd.read_csv("./data/crosswalk/optionmetrics_crsp_crosswalk.csv")
df_firms['SecurityID'] = df_firms['SecurityID'].astype('Int64')
df_firms = df_firms.merge(df_option_crsp, how = 'left', left_on = 'SecurityID', right_on = 'secid')
```

```python
## merge with Compustats
df_crsp_compustats = pd.read_csv("./data/crosswalk/crsp_compustats_crosswalk.csv", low_memory=False)
df_ccm_link = df_crsp_compustats[['gvkey', 'conm', 'tic', 'cusip', 'sic', 'naics', 'LPERMNO', 'EIN']].drop_duplicates()
df_ccm_link = df_ccm_link.loc[df_ccm_link['LPERMNO'].notna(), :]

df_firms = df_firms.merge(df_ccm_link, how = 'left', left_on = 'PERMNO', right_on = 'LPERMNO')
```
